```python
annees_plot = [2009, 2010, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021] # les années où les EPCI sont présents
plt.figure(figsize = (5, 3))
plt.scatter(annees_plot, nb_epci, marker = '*', color  = 'rebeccapurple', s = 75)
plt.title("Nombre d'EPCI dans les bilans annuels")
plt.xlabel("année")
plt.ylabel("nombre d'EPCI")
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```
